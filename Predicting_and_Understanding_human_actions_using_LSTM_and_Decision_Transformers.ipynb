```python
import cv2
import numpy as np

def mask_jersey_color(image, color):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    if color == "blue":
        lower = np.array([90, 80, 40])
        upper = np.array([130, 255, 255])
    elif color == "white":
        lower = np.array([0, 0, 200])
        upper = np.array([180, 40, 255])
    else:
        raise ValueError("Color must be 'blue' or 'white'")

    mask = cv2.inRange(hsv, lower, upper)
    return mask
```

```python
def extract_positions(mask):
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    centers = []
    for cnt in contours:
        if cv2.contourArea(cnt) > 100:  # filter noise
            M = cv2.moments(cnt)
            if M["m00"] != 0:
                cx = int(M["m10"] / M["m00"])
                cy = int(M["m01"] / M["m00"])
                centers.append((cx, cy))
    return sorted(centers)  # sorted left to right
```

```python
def mask_jersey_color(image, color):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    if color == "blue":
        lower = np.array([90, 80, 40])
        upper = np.array([130, 255, 255])
    elif color == "white":
        lower = np.array([0, 0, 200])
        upper = np.array([180, 40, 255])
    else:
        raise ValueError("Color must be 'blue' or 'white'")

    mask = cv2.inRange(hsv, lower, upper)
    return mask
```

```python
import cv2
import numpy as np

def extract_player_positions(image, color):
    mask = mask_jersey_color(image, color)

    # Find contours from mask
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    positions = []
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area > 100:  # Filter out small noise
            M = cv2.moments(cnt)
            if M["m00"] != 0:
                cx = int(M["m10"] / M["m00"])
                cy = int(M["m01"] / M["m00"])
                positions.append((cx, cy))

    return positions
```

```python
def pad_or_trim(players, num_players=3):
    """
    Ensures a fixed number of players in each frame by padding with (0,0) or trimming.
    """
    if len(players) > num_players:
        return players[:num_players]
    elif len(players) < num_players:
        return players + [(0, 0)] * (num_players - len(players))
    else:
        return players
```

```python
blue_team_positions = []
white_team_positions = []

frame_files = sorted([f for f in os.listdir(frames_dir) if f.endswith(".png")])

for fname in frame_files:
    img_path = os.path.join(frames_dir, fname)
    img = cv2.imread(img_path)

    blue_pos = pad_or_trim(extract_player_positions(img, "blue"))
    white_pos = pad_or_trim(extract_player_positions(img, "white"))

    blue_team_positions.append(blue_pos)
    white_team_positions.append(white_pos)
```

```python
import numpy as np

def compute_motion_features(positions):
    """
    positions: list of [(x1, y1), (x2, y2), ...] for each frame
    returns: np.array of shape (num_frames, num_players, 6)
    """
    positions = np.array(positions)  # (frames, players, 2)
    velocities = np.diff(positions, axis=0, prepend=positions[0:1])
    accelerations = np.diff(velocities, axis=0, prepend=velocities[0:1])

    return np.concatenate([positions, velocities, accelerations], axis=-1)
```

```python
def create_sequences(data, seq_len=10):
    """
    data: np.array of shape (frames, features)
    returns: X shape (num_samples, seq_len, features), Y shape (num_samples, features)
    """
    X, Y = [], []
    for i in range(len(data) - seq_len):
        X.append(data[i:i+seq_len])
        Y.append(data[i+seq_len])  # predict next frame
    return np.array(X), np.array(Y)
```

```python
import torch
import torch.nn as nn

class LSTMDecisionTransformer(nn.Module):
    def __init__(self, input_dim=36, embed_dim=64, lstm_hidden=64, ff_dim=128, seq_len=10, output_dim=36):
        super(LSTMDecisionTransformer, self).__init__()

        self.lstm = nn.LSTM(input_dim, lstm_hidden, batch_first=True)
        self.input_proj = nn.Linear(lstm_hidden, embed_dim)
        self.pos_embed = nn.Parameter(torch.zeros(1, seq_len, embed_dim))

        encoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=4, dim_feedforward=ff_dim)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=2)

        self.output = nn.Sequential(
            nn.Linear(embed_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 12)
        )

    def forward(self, x):
        # x shape: (batch, seq_len, input_dim)
        lstm_out, _ = self.lstm(x)                             # (batch, seq_len, lstm_hidden)
        proj = self.input_proj(lstm_out) + self.pos_embed      # (batch, seq_len, embed_dim)
        transformer_out = self.transformer(proj)               # (batch, seq_len, embed_dim)
        return self.output(transformer_out[:, -1, :])          # predict next frame from last timestep
```

```python
import torch
import torch.nn as nn

class LSTMDecisionTransformer(nn.Module):
    def __init__(self, input_dim=36, embed_dim=64, lstm_hidden=64, ff_dim=128, seq_len=10, output_dim=12, dropout=0.3):
        super(LSTMDecisionTransformer, self).__init__()

        self.lstm = nn.LSTM(input_dim, lstm_hidden, batch_first=True)
        self.dropout1 = nn.Dropout(dropout)

        self.input_proj = nn.Linear(lstm_hidden, embed_dim)
        self.pos_embed = nn.Parameter(torch.zeros(1, seq_len, embed_dim))

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embed_dim,
            nhead=4,
            dim_feedforward=ff_dim,
            dropout=dropout,
            batch_first=True  # Optional, good for clarity
        )
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=2)

        self.output = nn.Sequential(
            nn.Linear(embed_dim, 64),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(64, output_dim)
        )

    def forward(self, x):
        lstm_out, _ = self.lstm(x)                       # (B, T, lstm_hidden)
        lstm_out = self.dropout1(lstm_out)

        x = self.input_proj(lstm_out) + self.pos_embed   # (B, T, embed_dim)
        x = self.transformer(x)                          # (B, T, embed_dim)

        return self.output(x[:, -1, :])                  # (B, output_dim)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np

# 1. Define Model
class LSTMDecisionTransformer(nn.Module):
    def __init__(self, input_dim=36, embed_dim=64, lstm_hidden=64, ff_dim=128, seq_len=10, output_dim=12, dropout=0.3):
        super(LSTMDecisionTransformer, self).__init__()
        self.lstm = nn.LSTM(input_dim, lstm_hidden, batch_first=True)
        self.dropout1 = nn.Dropout(dropout)
        self.input_proj = nn.Linear(lstm_hidden, embed_dim)
        self.pos_embed = nn.Parameter(torch.zeros(1, seq_len, embed_dim))
        encoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=4, dim_feedforward=ff_dim,
                                                   dropout=dropout, batch_first=True)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=2)
        self.output = nn.Sequential(
            nn.Linear(embed_dim, 64),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(64, output_dim)
        )
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        x = self.dropout1(lstm_out)
        x = self.input_proj(x) + self.pos_embed
        x = self.transformer(x)
        return self.output(x[:, -1, :])

# 2. Filter Function to Remove Noisy Frames
def is_valid_frame(player_positions, threshold=2):
    return sum(p == (0, 0) for p in player_positions) < threshold

# 3. Feature Construction
def compute_motion_features(positions):
    positions = np.array(positions)
    velocities = np.diff(positions, axis=0, prepend=positions[0:1])
    accelerations = np.diff(velocities, axis=0, prepend=velocities[0:1])
    return np.concatenate([positions, velocities, accelerations], axis=-1)

# 4. Sequence Builder
def create_sequences(data, seq_len=10):
    X, Y = [], []
    for i in range(len(data) - seq_len):
        X.append(data[i:i+seq_len])
        Y.append(data[i+seq_len][:12])  # Predict only x, y (12 values)
    return np.array(X), np.array(Y)

# 5. Normalize and Prepare Dataset
def prepare_data(all_features):
    X, Y = create_sequences(all_features, seq_len=10)
    scaler = StandardScaler()
    X_flat = X.reshape(-1, X.shape[-1])
    scaler.fit(X_flat)
    X_scaled = scaler.transform(X_flat).reshape(X.shape)
    Y_scaled = scaler.transform(np.hstack([Y, np.zeros((Y.shape[0], 24))]))[:, :12]
    return X_scaled, Y_scaled, scaler

# 6. Train Model
def train_model(X_scaled, Y_scaled, device='cuda' if torch.cuda.is_available() else 'cpu'):
    X_train, X_val, Y_train, Y_val = train_test_split(X_scaled, Y_scaled, test_size=0.2, shuffle=False)
    train_loader = DataLoader(TensorDataset(torch.tensor(X_train, dtype=torch.float32),
                                            torch.tensor(Y_train, dtype=torch.float32)), batch_size=32, shuffle=True)
    val_loader = DataLoader(TensorDataset(torch.tensor(X_val, dtype=torch.float32),
                                          torch.tensor(Y_val, dtype=torch.float32)), batch_size=32)

    model = LSTMDecisionTransformer().to(device)
    criterion = nn.L1Loss()
    optimizer = optim.Adam(model.parameters(), lr=1e-4)
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=30, gamma=0.5)

    for epoch in range(100):
        model.train()
        total_train_loss = 0
        for X_batch, Y_batch in train_loader:
            X_batch, Y_batch = X_batch.to(device), Y_batch.to(device)
            optimizer.zero_grad()
            preds = model(X_batch)
            loss = criterion(preds, Y_batch)
            loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            total_train_loss += loss.item()
        scheduler.step()
        # Validation
        model.eval()
        total_val_loss = 0
        with torch.no_grad():
            for X_batch, Y_batch in val_loader:
                X_batch, Y_batch = X_batch.to(device), Y_batch.to(device)
                preds = model(X_batch)
                total_val_loss += criterion(preds, Y_batch).item()
                max_possible_loss = 6000  # (640 + 360) * 6 players
                norm_train_loss = total_train_loss / max_possible_loss
                norm_val_loss = total_val_loss / max_possible_loss
            print(f"Epoch {epoch+1}/100 | 🔁 Normalized Train Loss: {norm_train_loss:.4f} | 🧪 Val Loss: {norm_val_loss:.4f}")
    return model
```

```python
import torch
import cv2
import os
import numpy as np
import joblib
from sklearn.metrics import mean_absolute_error, r2_score

# Ensure consistent player input (3 red + 3 yellow)
def pad_or_trim(positions, num=3):
    positions = positions[:num]
    while len(positions) < num:
        positions.append((0, 0))
    return positions

def compute_motion_features(positions):
    pos = np.array(positions)
    vel = np.diff(pos, axis=0, prepend=pos[0:1])
    acc = np.diff(vel, axis=0, prepend=vel[0:1])
    return np.concatenate([pos, vel, acc], axis=-1)  # shape: (frames, 6, 6)

def create_sequences(data, seq_len=10):
    X, Y = [], []
    for i in range(len(data) - seq_len):
        X.append(data[i:i+seq_len])
        Y.append(data[i+seq_len][:12])  # predict only x, y
    return np.array(X), np.array(Y)
```

```python
def mask_jersey_color(image, color):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    if color == "red":
        lower1 = np.array([0, 100, 100])
        upper1 = np.array([10, 255, 255])
        lower2 = np.array([160, 100, 100])
        upper2 = np.array([179, 255, 255])
        mask = cv2.inRange(hsv, lower1, upper1) | cv2.inRange(hsv, lower2, upper2)

    elif color == "yellow":
        lower = np.array([20, 100, 100])
        upper = np.array([30, 255, 255])
        mask = cv2.inRange(hsv, lower, upper)

    else:
        raise ValueError("Unsupported jersey color")

    return mask
```

```python
def extract_player_positions(image, color):
    mask = mask_jersey_color(image, color)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    positions = []
    for cnt in contours:
        if cv2.contourArea(cnt) > 100:
            M = cv2.moments(cnt)
            if M["m00"] != 0:
                cx = int(M["m10"] / M["m00"])
                cy = int(M["m01"] / M["m00"])
                positions.append((cx, cy))
    return positions
```

```python
red_positions = []
yellow_positions = []

for fname in sorted(os.listdir("test_frames")):
    img = cv2.imread(os.path.join("test_frames", fname))
    red = pad_or_trim(extract_player_positions(img, "red"))
    yellow = pad_or_trim(extract_player_positions(img, "yellow"))

    if sum(p == (0, 0) for p in red + yellow) < 4:
        red_positions.append(red)
        yellow_positions.append(yellow)
```

```python
import cv2
import numpy as np

def mask_jersey_color(image, color):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    if color == "red":
        # Adjust HSV ranges if needed
        lower1 = np.array([0, 50, 50])  # Lowered saturation and value
        upper1 = np.array([10, 255, 255])
        lower2 = np.array([170, 50, 50])  # Lowered saturation and value
        upper2 = np.array([180, 255, 255])
        mask = cv2.inRange(hsv, lower1, upper1) | cv2.inRange(hsv, lower2, upper2)

    elif color == "yellow":
        # Adjust HSV ranges if needed
        lower = np.array([20, 100, 100])
        upper = np.array([40, 255, 255])  # Widened hue range
        mask = cv2.inRange(hsv, lower, upper)

    else:
        raise ValueError("Unsupported jersey color")

    return mask

def extract_player_positions(image, color):
    mask = mask_jersey_color(image, color)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    positions = []
    for cnt in contours:
        # Experiment with area threshold (e.g., 50, 75, 100)
        if cv2.contourArea(cnt) > 50:  # Adjusted area threshold
            M = cv2.moments(cnt)
            if M["m00"] != 0:
                cx = int(M["m10"] / M["m00"])
                cy = int(M["m01"] / M["m00"])
                positions.append((cx, cy))
    return positions
```

```python
# Create (x, y, vx, vy, ax, ay) for all players
red_feat = compute_motion_features(red_positions)
yellow_feat = compute_motion_features(yellow_positions)
all_features = np.concatenate([red_feat, yellow_feat], axis=1).reshape(len(red_feat), -1)

# Sequence it
X_test, Y_test_actual = create_sequences(all_features)

# Load scaler and normalize
scaler = joblib.load("scaler.pkl")
X_test_scaled = scaler.transform(X_test.reshape(-1, X_test.shape[-1])).reshape(X_test.shape)

# Pad Y for scaling (required shape)
Y_test_pad = np.hstack([Y_test_actual, np.zeros((Y_test_actual.shape[0], 24))])
Y_test_scaled = scaler.transform(Y_test_pad)[:, :12]
```

```python
import torch
import torch.nn as nn

class LSTMDecisionTransformer(nn.Module):
    def __init__(self, input_dim=36, embed_dim=64, lstm_hidden=64, ff_dim=128, seq_len=10, output_dim=12, dropout=0.3):
        super(LSTMDecisionTransformer, self).__init__()

        self.lstm = nn.LSTM(input_dim, lstm_hidden, batch_first=True)
        self.dropout1 = nn.Dropout(dropout)

        self.input_proj = nn.Linear(lstm_hidden, embed_dim)
        self.pos_embed = nn.Parameter(torch.zeros(1, seq_len, embed_dim))

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embed_dim,
            nhead=4,
            dim_feedforward=ff_dim,
            dropout=dropout,
            batch_first=True  # Optional, good for clarity
        )
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=2)

        self.output = nn.Sequential(
            nn.Linear(embed_dim, 64),
            nn.ReLU(),
            # Removed Dropout layer
            nn.Linear(64, output_dim)
        )

    def forward(self, x):
        lstm_out, _ = self.lstm(x)                       # (B, T, lstm_hidden)
        lstm_out = self.dropout1(lstm_out)

        x = self.input_proj(lstm_out) + self.pos_embed   # (B, T, embed_dim)
        x = self.transformer(x)                          # (B, T, embed_dim)

        return self.output(x[:, -1, :])                  # (B, output_dim)
```

```python
model = LSTMDecisionTransformer()
model.load_state_dict(torch.load("hybrid_model_best.pth", map_location="cpu"))
model.eval()

with torch.no_grad():
    X_tensor = torch.tensor(X_test_scaled, dtype=torch.float32)
    preds_scaled = model(X_tensor).cpu().numpy()
```

```python
# ... (previous code for loading model, preprocessing, prediction, etc.) ...

# Inverse scale to pixel space
pad_preds = np.hstack([preds_scaled, np.zeros((preds_scaled.shape[0], 24))])
pad_truth = np.hstack([Y_test_scaled, np.zeros((Y_test_scaled.shape[0], 24))])

preds_actual = scaler.inverse_transform(pad_preds)[:, :12]
truths_actual = scaler.inverse_transform(pad_truth)[:, :12]

# Calculate MAE
mae = mean_absolute_error(truths_actual, preds_actual)
norm_mae = mae / 6000  # max possible loss

# Calculate RMSE
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(truths_actual, preds_actual))
norm_rmse = rmse / 6000  # max possible loss


print(f"📉 Normalized MAE: {norm_mae:.4f}")
print(f"📉 Normalized RMSE: {norm_rmse:.4f}")
```
